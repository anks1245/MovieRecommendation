```python
movies_ds.dropna(inplace=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
cv.fit_transform(movies_ds['Overview']).toarray().shape
```

```python
vectors = cv.fit_transform(movies_ds['Overview']).toarray()
```

```python
vectors[0]
```

```python
len(cv.get_feature_names_out())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies_ds['Overview'] = movies_ds['Overview'].apply(stem)
```

```python
movies_ds['Overview']
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommendation(movie):
    movie_index = movies_ds[movies_ds['Series_Title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies_ds.iloc[i[0]].Series_Title)
```

```python
recommendation("The Godfather")
```
